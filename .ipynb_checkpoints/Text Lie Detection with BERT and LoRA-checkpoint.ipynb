# Project Overview

M2D2 (Multimodal Deception Detection Model) is a transformer-based framework for detecting deception in audio-visual data. Using the DOLOS dataset (1,675 clips, 213 subjects), it implements trimodal fusion of visual (ViT-B/16), audio (Wav2Vec2-base), and text (BERT-base via Whisper ASR) modalities with cross-attention fusion, targeting 75-85% accuracy. The system employs lightweight Uniform Temporal Adapters for parameter-efficient fine-tuning (5-10% of 300M parameters).

**Keywords**: Deception Detection, Multimodal Learning, Transformer Networks, Cross-Modal Fusion

---


## Dataset: DOLOS

- **Clips**: 1,675 videos (2-19s, median 5s)
- **Subjects**: 213 (141 male, 72 female)
- **Labels**: Binary (truthful/deceptive)
- **Annotations**: MUMIN scheme (25 facial + 5 speech cues)
- **Splits**: 70% train (1,172) | 15% val (251) | 15% test (252)

**Preprocessing**: Frame extraction (30 FPS), audio resampling (16kHz), ASR transcription (Whisper), tokenization (BERT, max len 512)

# Lie vs. Truth Text Classification with BERT + LoRA

This project trains a binary text classifier to detect **lying vs. truthful** statements using **BERT** with **LoRA fine‑tuning**.

- Label **truth**: files whose names contain `truth`
- Label **lie**: files whose names contain `lie`
- Data directory: `../data/DOLOS/txt_files/txt_final`
- Model: `bert-base-uncased` + LoRA adapters

## TXT Project Structure

Example layout:

```text
project_root/
├─ train_bert_lie_truth.py
└─ ../data/DOLOS/txt_files/txt_final/
   ├─ AN_WILTY_EP15_lie11.txt
   ├─ AN_WILTY_EP15_lie12.txt
   ├─ AN_WILTY_EP15_truth1.txt
   ├─ AN_WILTY_EP15_truth2.txt
   └─ ...
```
Each .txt file contains one sample (dialogue / transcript segment).

## Data Format
Directory: ../data/DOLOS/txt_files/txt_final

Label is derived from the filename:

If filename (lowercased) contains "lie" → label 1 (lie)
If filename (lowercased) contains "truth" → label 0 (truth)
Files without lie or truth in the name are ignored
Example:

```text
AN_WILTY_EP15_lie11.txt    → label = 1 (lie)
AN_WILTY_EP15_truth3.txt   → label = 0 (truth)
```
## Training Script
The main training script is:

```text
train_bert_lie_truth.py
```

It performs:
Load all .txt files under ../data/DOLOS/txt_files/txt_final
Assign labels based on filename (lie / truth)
Split into train / validation (80% / 20%, stratified)
Tokenize with bert-base-uncased
Fine‑tune BERT with LoRA adapters as a binary classifier
Save the fine‑tuned model and tokenizer
Run:
```text
python train_bert_lie_truth.py
```
Key training settings (typical):
Model: bert-base-uncased
LoRA: rank r = 8, lora_alpha = 16, lora_dropout = 0.1 on attention query and value modules
Epochs: e.g. num_train_epochs = 5 (or higher, e.g. 50, for small datasets)
Batch size: e.g. per_device_train_batch_size = 16
Learning rate: 2e-4 for LoRA (or 2e-5 for full fine‑tuning)
The script prints training progress and final metrics on the validation set.

## Inference / Prediction
After training, the model is saved to:
```text
./bert_lie_truth_lora
```
A minimal inference example:

```bash
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from peft import PeftModel

MODEL_DIR = "./bert_lie_truth_lora"
BASE_MODEL_NAME = "bert-base-uncased"

tokenizer = BertTokenizer.from_pretrained(MODEL_DIR)
base_model = BertForSequenceClassification.from_pretrained(
    BASE_MODEL_NAME,
    num_labels=2
)
model = PeftModel.from_pretrained(base_model, MODEL_DIR)
model.eval()

def predict(text: str) -> str:
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=256
    )
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        pred = torch.argmax(logits, dim=-1).item()
    return "lie" if pred == 1 else "truth"

example = "This is an example statement."
print(predict(example))
```
## Notes

### WAV transform to TXT

````python
import os
import whisper

model = whisper.load_model("medium")  # medium / large-v3

audio_dir = "./audio_files"
out_dir = "./txt_files/txt_M"
os.makedirs(out_dir, exist_ok=True)

for name in os.listdir(audio_dir):
    if not name.lower().endswith(".wav"):
        continue
    path = os.path.join(audio_dir, name)
    print("Transcribing:", path)
    result = model.transcribe(path, language="en")  # language="zh"
    txt_path = os.path.join(out_dir, os.path.splitext(name)[0] + ".txt")
    with open(txt_path, "w", encoding="utf-8") as f:
        f.write(result["text"])
````

````python
import os
import whisper

model = whisper.load_model("large-v3")  # medium / large-v3

audio_dir = "./audio_files"
out_dir = "./txt_files/txt_LV3"
os.makedirs(out_dir, exist_ok=True)

for name in os.listdir(audio_dir):
    if not name.lower().endswith(".wav"):
        continue
    path = os.path.join(audio_dir, name)
    print("Transcribing:", path)
    result = model.transcribe(path, language="en")  # language="zh"
    txt_path = os.path.join(out_dir, os.path.splitext(name)[0] + ".txt")
    with open(txt_path, "w", encoding="utf-8") as f:
        f.write(result["text"])
````

### Compared the file txt_M and the file txt_LV3

````python
import os
import filecmp
import shutil
import difflib

def show_file_diff(file1, file2, filename):
    print(f"--- begin compared: {filename} ---")
    with open(file1, 'r', encoding='utf-8') as f1, open(file2, 'r', encoding='utf-8') as f2:
        lines1 = f1.readlines()
        lines2 = f2.readlines()

        diff = difflib.unified_diff(
            lines1,
            lines2,
            fromfile=file1,
            tofile=file2,
            lineterm=''
        )

        for line in diff:
            print(line)
    print(f"--- file {filename} compared finish ---\n")


def compare_txt_dirs(dir_a, dir_b):
    txts_a = {f for f in os.listdir(dir_a) if f.endswith('.txt')}
    txts_b = {f for f in os.listdir(dir_b) if f.endswith('.txt')}

    common_txts = txts_a & txts_b
    only_in_a = txts_a - txts_b
    only_in_b = txts_b - txts_a

    print("TXT files only in A:", only_in_a)
    print("TXT files only in B:", only_in_b)

    all_equal = True

    for name in sorted(common_txts):
        path_a = os.path.join(dir_a, name)
        path_b = os.path.join(dir_b, name)

        same = filecmp.cmp(path_a, path_b, shallow=False)

        if same:
            print(f"[SAME] {name}")
        else:
            all_equal = False
            print(f"[DIFFERENT] {name}")
            show_file_diff(path_a, path_b, name)

    if all_equal and not only_in_a and not only_in_b:
        print("All txt files with the same name are identical and no extra/missing txt files.")
    else:
        print("compared finished.")
````

````python
dir_a = r"./txt_files/txt_M"
dir_b = r"./txt_files/txt_LV3"
compare_txt_dirs(dir_a, dir_b)
````

### Train the BERT model with LoRA

````python
import os
from sklearn.model_selection import train_test_split
import numpy as np
import torch
from torch.utils.data import Dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

from transformers import (
    BertTokenizer,
    BertForSequenceClassification,
    Trainer,
    TrainingArguments,
)
from peft import LoraConfig, get_peft_model, TaskType

DATA_DIR = r"./txt_files/txt_LV3"
MODEL_NAME = "bert-base-uncased"
OUTPUT_DIR = "./bert_lie_truth_lora"

# 1. load data
texts = []
labels = []  # 0 = truth, 1 = lie

for fname in os.listdir(DATA_DIR):
    if not fname.endswith(".txt"):
        continue
    lower_name = fname.lower()
    if "lie" in lower_name:
        label = 1
    elif "truth" in lower_name:
        label = 0
    else:
        continue

    path = os.path.join(DATA_DIR, fname)
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read().strip()
    if not text:
        continue

    texts.append(text)
    labels.append(label)

print(f"Loaded {len(texts)} samples.")

X_train, X_val, y_train, y_val = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)

tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)

class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=256):
        encodings = tokenizer(
            texts,
            truncation=True,
            padding=True,
            max_length=max_len
        )
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = TextDataset(X_train, y_train, tokenizer)
val_dataset = TextDataset(X_val, y_val, tokenizer)

# 2. based model
base_model = BertForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=2
)

# 3. config LoRA
lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,  
    r=8,                        
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=["query", "value"]  
)

model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters() 

# 4. evaul func
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average="binary", pos_label=1
    )
    return {
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }

# 5. train

training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    num_train_epochs=50,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    learning_rate=2e-4,  
    weight_decay=0.01,
    logging_steps=10,
    fp16=True,           
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)
````

````python
trainer.train()
````

````python
# save LoRA + base bert
trainer.save_model(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)
print(f"LoRA model saved to {OUTPUT_DIR}")
````

### Predict the content by trained BERT model with LoRA

````python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from peft import PeftModel

MODEL_DIR = "./bert_lie_truth_lora"
BASE_MODEL_NAME = "bert-base-uncased"

tokenizer = BertTokenizer.from_pretrained(MODEL_DIR)
base_model = BertForSequenceClassification.from_pretrained(
    BASE_MODEL_NAME, num_labels=2
)
model = PeftModel.from_pretrained(base_model, MODEL_DIR)
model.eval()

def predict(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=256
    )
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        pred = torch.argmax(logits, dim=-1).item()
    return "lie" if pred == 1 else "truth"

print(predict("The only concert I have ever been to was by show.")) # the content from file AN_WILTY_EP21_lie18.txt
print(predict("Yeah, well, I would think that if people didn't think they were...")) # the content from file N_WILTY_EP20_truth4.txt
````
